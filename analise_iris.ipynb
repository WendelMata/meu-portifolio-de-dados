```python
import seaborn as sns
import pandas as pd

df = sns.load_dataset('iris')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Configurar o estilo dos grficos
sns.set(style="whitegrid")

# Grfico de distribuio do comprimento das ptalas
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x="petal_length", hue="species", kde=True)
plt.title("Distribuio do Comprimento das Ptalas por Espcie")
plt.show()
```

```python
# Grfico de disperso entre comprimento e largura das ptalas
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x="petal_length", y="petal_width", hue="species", palette="Set1")
plt.title("Relao entre Comprimento e Largura das Ptalas")
plt.show()
```

```python
# Importar bibliotecas necessrias
import matplotlib.pyplot as plt
import seaborn as sns

# Calcular a matriz de correlao apenas com colunas numricas
correlation_matrix = df.drop(columns=["species"]).corr()

# Plotar a matriz de correlao
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Matriz de Correlao")
plt.show()
```
