```python
import seaborn as sns
import pandas as pd

df = sns.load_dataset('iris')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Configurar o estilo dos gráficos
sns.set(style="whitegrid")

# Gráfico de distribuição do comprimento das pétalas
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x="petal_length", hue="species", kde=True)
plt.title("Distribuição do Comprimento das Pétalas por Espécie")
plt.show()
```

```python
# Gráfico de dispersão entre comprimento e largura das pétalas
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x="petal_length", y="petal_width", hue="species", palette="Set1")
plt.title("Relação entre Comprimento e Largura das Pétalas")
plt.show()
```

```python
# Importar bibliotecas necessárias
import matplotlib.pyplot as plt
import seaborn as sns

# Calcular a matriz de correlação apenas com colunas numéricas
correlation_matrix = df.drop(columns=["species"]).corr()

# Plotar a matriz de correlação
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Matriz de Correlação")
plt.show()
```
